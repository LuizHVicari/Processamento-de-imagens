# Exemplos OpenCv

````python
import cv2
````

## Leitura de Imagem

````python
img = cv2.imread('lena.png')
````

```
cv2.imread('nome_da_imagem.jpg', 0)
```

Para imagens coloridas:

```

```

```

```

## Exibindo a imagem

````python
import matplotlib.pyplot as plt

plt.imshow(img);
````

O OpenCV usa o formato BGR enquanto o matplotlib usa o formato RGB, portanto é necessário converter

````python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
````

Criando um degrade para a imagem

````python
for y in range(0, img.shape[0]):
    for x in range(0, img.shape[1]):
        img[y, x] = (x%256, y%256, x%256)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
for y in range(0, img.shape[0]):
    for x in range(0, img.shape[1]):
        img[y, x] = (0, (x*y)%256, 0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()   
````

````python
for y in range(0, img.shape[0], 10):
    for x in range(0, img.shape[1], 10):
        img[y:y+5, x:x+5] = (0, 255, 255)
        img[y+5:y+10, x+5:x+10] = (0, 255, 255)
        img[y:y+5, x+5:x+10] = (255, 0, 0)
        img[y+5:y+10, x:x+5] = (255, 0, 0)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()   
````

É necessário ler novamente a imagem para reverter ao original, poisa variável sobrecarrega as modificações em cima dela

````python
img = cv2.imread('lena.png')

for y in range(0, img.shape[0], 10):
    for x in range(0, img.shape[1], 10):
        img[y:y+5, x:x+5] = (0,255,255)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

## Salvando a Imagem

O método 

```
cv2.imwrite('nome_do_arquivo')
```

salva a imagem

## Redimensionament da imagem

````python
img = cv2.imread('lena.png')

largura = img.shape[1]
altura = img.shape[0]

proporcao = float(altura / largura)

largura_nova = 520
altura_nova = int(largura_nova * proporcao)
tamanho_novo = (largura_nova, altura_nova)

img_redimensionada = cv2.resize(img, tamanho_novo, interpolation = cv2.INTER_AREA)

plt.imshow(cv2.cvtColor(img_redimensionada, cv2.COLOR_BGR2RGB))
plt.show()   
````

Para girar uma imagem, o método

```
cv2.flip(img, num)
```

gira para:

    1: horizontal

    0: vertical

    -1: deixa de cabeça para baixo
