```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_img_rgb(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show() 
```

## Leitura da imagem

```python
img = cv2.imread('pinguim.jpg')
show_img_rgb(img)
```

## Mostre as dimens√µes (shape) e n√∫mero de canais

```python
print(f'A imagem tem {img.shape[0]} colunas, {img.shape[1]} linhas e {img.shape[2]} canais')
```

## Insira um pequeno ret√¢ngulo na posi√ß√£o 0,0, de dimens√£o (20,20)

```python
img_copy = img.copy()
img_copy[0:20, 0:20] = (0, 0, 0)
show_img_rgb(img_copy)
```

## Fa√ßa o quarto quadrante da imagem ficar pintada de preto.

```python
img_copy[int(img_copy.shape[0] / 2) : img_copy.shape[0], int(img_copy.shape[1] / 2) : img_copy.shape[1]] = (0, 0, 0)
show_img_rgb(img_copy)
```

## Rotacione a imagem em 75 graus

```python
(lar, alt) = img.shape[:2]

centro = (lar // 2, alt // 2)

M = cv2.getRotationMatrix2D(centro, 30, 1.0)
img_copy = cv2.warpAffine(img_copy, M, (lar, alt))
show_img_rgb(img_copy)
```

## Fa√ßa um flip horizontal na imagem

```python
img_copy = cv2.flip(img_copy, 1)
show_img_rgb(img_copy)
```

## Redimensione a imagem para largura de 200 pixels

```python
def resize_img_width(img, new_width):
    width, height = img.shape[:2]
    prop = height / width

    new_height = int(new_width * prop)
    new_size = (new_width, new_height)

    img_resized = cv2.resize(img, new_size, interpolation=cv2.INTER_AREA)

    return img_resized
```

```python
def show_img_rgb_axis(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show() 
```

```python
img_copy = resize_img_width(img=img_copy, new_width=200)
show_img_rgb_axis(img_copy)
```

## Salve a imagem no disco com o nome "nova_imagem.png"

```python
cv2.imwrite('nova_imagem.png', img_copy)
```
