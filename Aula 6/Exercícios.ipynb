## Exercício 1)

Os pixels de uma imagem são representados na matriz abaixo. As pequenas 
flutuações de intensidade caracterizam a presença de um ruído na imagem.

```python
import numpy as np

mat = np.array([
    [4, 5, 7, 7, 7, 8, 6],
    [7, 6, 7, 5, 7, 7, 7],
    [6, 5, 4, 10, 12, 12, 11],
    [10, 9, 8, 7, 5, 5, 6],
    [11, 8, 8, 8, 7, 6, 6],
    [5, 6, 7, 6, 6, 6, 6],
    [4, 5, 10, 9, 9, 8, 8]])
```

### A)
Existem pixels adjacentes? No sentido 4-P, 8P ou diagonal?

```python
def find_4p_adj(mat):
    adj = []
    for indice_linha, linha in enumerate(mat):
        for indice_coluna, coluna in enumerate(linha):
            try:
                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna] and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna))) 
                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna] and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna)))
                if mat[indice_linha][indice_coluna] == mat[indice_linha][indice_coluna + 1] and indice_coluna < mat.shape[1]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha, indice_coluna + 1)))
                if mat[indice_linha][indice_coluna] == mat[indice_linha][indice_coluna - 1] and indice_coluna > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha, indice_coluna - 1)))
            except:
                pass

    return adj

print(f'Os pontos adjacentes na diagonal são: {find_4p_adj(mat)}')
```

```python
def find_8p_adj(mat):
    adj = []
    for indice_linha, linha in enumerate(mat):
        for indice_coluna, coluna in enumerate(linha):
            try:
                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna] and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna] and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha][indice_coluna + 1] and indice_coluna < mat.shape[1]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha, indice_coluna+ 1)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha][indice_coluna - 1] and indice_coluna > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha, indice_coluna - 1)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna + 1] and indice_coluna < mat.shape[1] and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna + 1)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna - 1] and indice_coluna > 0 and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna - 1)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna - 1] and indice_coluna > 0 and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna - 1)))

                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna + 1] and indice_coluna < mat.shape[1] and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna + 1)))
                    
            except:
                pass

    return adj

print(f'Os pontos adjacentes na diagonal são: {find_8p_adj(mat)}')
```

```python
def find_diag_adj(mat):
    adj = []
    for indice_linha, linha in enumerate(mat):
        for indice_coluna, coluna in enumerate(linha):
            try:
                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna + 1] and indice_coluna < mat.shape[1] and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna + 1)))
                if mat[indice_linha][indice_coluna] == mat[indice_linha + 1][indice_coluna - 1] and indice_coluna > 0 and indice_linha < mat.shape[0]:
                    adj.append(((indice_linha, indice_coluna), (indice_linha + 1, indice_coluna - 1)))
                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna - 1] and indice_coluna > 0 and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna - 1)))
                if mat[indice_linha][indice_coluna] == mat[indice_linha - 1][indice_coluna + 1] and indice_coluna < mat.shape[1] and indice_linha > 0:
                    adj.append(((indice_linha, indice_coluna), (indice_linha - 1, indice_coluna + 1)))
            except:
                pass

    return adj

print(f'Os pontos adjacentes na diagonal são: {find_diag_adj(mat)}')
```

### B)
Calcule o Histograma de intensidades de níveis de cinza desta imagem 

```python
import seaborn as sns

ax = sns.displot(mat.flatten())
ax.set(title='Histograma de intensidades de níveis de cinza', xlabel='Intensidade de Cinza', ylabel='Quantidade de pixels')
ax;
```

### C)
Qual o histograma acumulado?

```python
ax = sns.histplot(mat.flatten(), cumulative=True)
ax.set(title='Histograma de intensidades de níveis de cinza acumulados', xlabel='Intensidade de Cinza', ylabel='Quantidade de pixels')
ax;
```

### D)
Qual o histograma normalizado?

```python
ax = sns.distplot(mat.flatten(), hist_kws = {'cumulative':True},
                  kde_kws = {'cumulative':True},)
ax.set(title='Histograma de intensidades de níveis de cinza acumulados e normalizados', xlabel='Intensidade de Cinza', ylabel='Quantidade de pixels')
ax.get_lines()[0].remove()
ax;
```

### E)
Para se determinar automaticamente a transformação de contraste a partir do histograma, podemos utilizar a fórmula Y = a*X + b,  que deve ser aplicada a uma imagem digital

```python
min = mat.min()
max = mat.max()

a = 255 / (max - min)
b = - a * min

x = lambda X : a * X + b

np.array(list(map(x, mat)))
```

### F)
Como fica a representação da imagem comprimida utilizando a seguinte transformação:

#### “intensidades <= 8” → 8

```python
x = lambda X : 8 if X <= 8 else X

np.array(list(map(x, mat.flatten()))).reshape(7, 7)
```

### “intensidades iguais a 9 ou 10” → 10

```python
x = lambda X : 10 if (X == 10 or X == 9) else X

np.array(list(map(x, mat.flatten()))).reshape(7, 7)
```

### “intensidades >= 11 “ → 11

```python
x = lambda X : 11 if X >= 11 else X

np.array(list(map(x, mat.flatten()))).reshape(7, 7)
```

## Exercício 2)
Implemente um programa que deverá solicitar ao usuário as coordenadas de dois pontos P1 e P2 (inicio e fim da região) localizados dentro dos limites do tamanho da  imagem  e  exibir  que  lhe  for  fornecida.  Entretanto,  a  região  definida  pelo retângulo definidos pelos pontos P1 e P2 será exibida com o negativo da imagem na região correspondente.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('lena.png')
plt.title('')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
x_i = int(input('Informe o x inicial: '))
x_f = int(input('Informe o x final: '))

if  x_i < 0 or x_i > img.shape[0] or x_f < 0 or x_f > img.shape[0]:
    raise Exception('Erro: os valores informados não estão presentes na imagem')

if x_i > x_f:
    x_i, x_f = x_f, x_i

y_i = int(input('Informe o y inicial: '))
y_f = int(input('Informe o y final: '))

if  y_i < 0 or y_i > img.shape[0] or y_f < 0 or y_f > img.shape[0]:
    raise Exception('Erro: os valores informados não estão presentes na imagem')

if y_i > y_f:
    y_i, y_f = y_f, y_i

img[x_i:x_f, y_i:y_f] = (255 - img[x_i:x_f, y_i:y_f])

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
