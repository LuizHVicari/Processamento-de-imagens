```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
kernel
```

## 1) Utilize o elemento estruturante abaixo:



### a) Aplique erosão e dilatação na imagem img1.png disponível no diretório. 

```python
img1 = cv2.imread('img1.png', cv2.IMREAD_GRAYSCALE)
img1_erode = cv2.erode(img1, kernel, iterations=1)
img1_dilate = cv2.dilate(img1, kernel, iterations=1)

plt.figure(figsize=(20, 20))

plt.subplot(3, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Imagem Original')

plt.subplot(3, 3, 2)
plt.imshow(img1_erode, cmap='gray')
plt.title('Imagem Erodida')

plt.subplot(3, 3, 3)
plt.imshow(img1_dilate, cmap='gray')
plt.title('Imagem Dilatada')

plt.show()
```

```python
np.sum(img1 == img1_erode) / (img1.shape[0] * img1.shape[1])
```

```python
np.sum(img1 == img1_dilate) / (img1.shape[0] * img1.shape[1])
```

### b) Aplique erosão, dilatação, fechamento na imagem img2.png

```python
img2 = cv2.imread('img2.png',cv2.IMREAD_GRAYSCALE)
img2_erode = cv2.erode(img2, kernel, iterations=1)
img2_dilate = cv2.dilate(img2, kernel, iterations=1)
img2_closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(20, 20))

plt.subplot(3, 4, 1)
plt.imshow(img2, cmap='gray')
plt.title('Imagem Original')

plt.subplot(3, 4, 2)
plt.imshow(img2_erode, cmap='gray')
plt.title('Imagem Erodida')

plt.subplot(3, 4, 3)
plt.imshow(img2_dilate, cmap='gray')
plt.title('Imagem Dilatada')

plt.subplot(3, 4, 4)
plt.imshow(img2_closing, cmap='gray')
plt.title('Imagem Closing')
plt.show()
```

## 2) Considere a imagem img4.png

### a) Aplique o filtro morfológigo de erosão (3x3). Qual é o efeito na imagem?

```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img4, cmap='gray')
```

### b) Aplique o filtro morfológigo de dilatação (3x3). Qual é o efeito na imagem?

```python
kernel = np.ones((3, 3))

img4_dilate = cv2.dilate(img4, kernel, iterations=1)

plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title('Imagem original')
plt.imshow(img4, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Imagem dilatada')
plt.imshow(img4_dilate, cmap='gray')
```

O efeito na imagem após a dilatação foi de reduzir o "granulado" nos objetos.

### c) Aplique o filtro morfológigo de abertura (3x3). Qual é o efeito na imagem?


```python
kernel = np.ones((3, 3))

img4_opening = cv2.morphologyEx(img4, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title('Imagem original')
plt.imshow(img4, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Imagem "aberta"')
plt.imshow(img4_opening, cmap='gray')
```

O efeito na imagem com a abertura foi de diminuir (ou eliminar) as partes brancas de pequenos tamanhos, assim quase retirando os objetos que possuiam o granulado dentro. 

### d) Aplique o filtro morfológigo de fechamento (3x3). Qual é o efeito na imagem?

```python
kernel = np.ones((3, 3))

img4_opening = cv2.morphologyEx(img4, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title('Imagem original')
plt.imshow(img4, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Imagem "fechada"')
plt.imshow(img4_opening, cmap='gray')
```

O efeito do fechamento na imagem foi de preencher os espaços vazios pretos dos objetos aumentando o preto e diminuindo o branco.

## 3) Como podemos obter as bordas dos objetos na imagem img5.png?

```python
img5 = cv2.imread('img5.png', cv2.IMREAD_GRAYSCALE)
img5_threshold = cv2.adaptiveThreshold(img5,
                                       255, 
                                       cv2.ADAPTIVE_THRESH_MEAN_C, 
                                       cv2.THRESH_BINARY, 
                                       11, 
                                       2)
img5_threshold = np.absolute(255 - img5_threshold)

img5_threshold = cv2.erode(img5_threshold, 
                           cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3)), 
                           iterations=2)

plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.imshow(img5, cmap='gray')
plt.title('Imagem original')


plt.subplot(2, 2, 2)
plt.imshow(img5_threshold, cmap='gray')
plt.title('Contornos da imagem')

plt.show()
```
