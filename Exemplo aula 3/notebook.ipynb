```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('lena.png')
show_img(img)
```

```python
img.shape
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
img[127:128, 127:128] = (255, 255, 255)
show_img(img)
```

```python
img[100:150, 50:100] = (0, 0, 255)
show_img(img)
```

```python
img = cv2.rectangle(img, (1, 1), (224, 224), (255, 0, 0))
show_img(img)
```

```python
cv2.imwrite('saida.jpg', img)
```
